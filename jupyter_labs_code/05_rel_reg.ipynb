```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
sns.scatterplot(x='total_bill',y='tip',
               data=tips,hue='sex')
```

```python
sns.scatterplot(x='total_bill',y='tip',
               data=tips,hue='sex',size='smoker')
```

```python
sns.lineplot(x=flights.month,y=flights.passengers,estimator=sum)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,col='sex',hue='smoker',size='time')
```

```python
sns.regplot(x='total_bill',y='tip',ci=False,
           data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',
          data=tips,ci=False,hue='smoker',
          col='day')
```

```python
sns.residplot(x='total_bill',y='tip',data=tips)
```

```python
fgrup = flights.pivot('month','year','passengers')
```

```python
sns.heatmap(fgrup)
```

```python
cmat = tips.corr()
```

```python
sns.heatmap(cmat,annot=True)
```
