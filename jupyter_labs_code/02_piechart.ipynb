```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
plt.isinteractive()
```

```python
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
daywise_bill = df.groupby('day',as_index= False).total_bill.sum()
daywise_bill
```

# bar garph and line graph

```python
plt.plot(daywise_bill.day,daywise_bill.total_bill);

plt.title("Day Wise Sales");
plt.xlabel("Day");
plt.ylabel("Total Bill")
```

```python
plt.figure(figsize=(8,4))
plt.bar(daywise_bill.day,daywise_bill.total_bill);
plt.title("Day Wise Sales");
plt.xlabel("Day");
plt.ylabel("Total Bill")
```

# Pie Chart

```python
plt.style.use('seaborn')
```

```python
plt.pie(daywise_bill.total_bill,
        labels = daywise_bill.day,shadow=True,
        labeldistance=0.75,autopct="%0.2f"
       ,explode=(0,0,0.05,0.05));
```

```python
plt.scatter(df.day,df.total_bill);

plt.title("Day Wise Sales");
plt.xlabel("Day");
plt.ylabel("Total Bill");
```

```python
plt.scatter(df.tip,df.total_bill);

plt.title("Day Wise Sales");
plt.xlabel("Day");
plt.ylabel("Total Bill");
```

```python
ax = plt.subplot();
ax.bar(daywise_bill.day,daywise_bill.total_bill);
ax.set_title("Day Wise Sales Data");
ax.set_xlabel("Day");
ax.set_ylabel("Total Bill");
```
