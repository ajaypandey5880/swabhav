```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips
```

```python
flights
```

```python
sns.boxplot(x='day',y='tip',data=tips)
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='day',y='tip',
            hue='sex',data=tips)
plt.gcf().savefig('abc.png')
```

```python
sns.countplot(x='day',data=tips,hue='sex')
```

```python
sns.barplot(x='day',y='tip',estimator=np.sum,
            data=tips,ci=False)
```

```python
sns.barplot(x='day',y='tip',estimator=np.sum,
            data=tips,ci=False,hue='sex')
```

```python
sns.catplot(x='sex',y='tip',data=tips,kind='bar',
           ci=False)
```

```python
sns.catplot(x='sex',y='tip',data=tips,kind='bar',
           ci=False,col='day')
```

```python
plt.figure(figsize=(15,5))
plt.xticks(rotation=60)
sns.barplot(x='month',y='passengers',data=flights,
           ci=False);
```

```python
plt.figure(figsize=(15,5))
sns.barplot(y='month',x='passengers',data=flights,
           ci=False);
```
